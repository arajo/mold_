# 2번, 3번 file 분 단위로 join

```python
import pandas as pd
import numpy as np
import datetime

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
file2 = pd.read_csv('./data/isee_machine_data_20180901_to_20190331_C123.csv', encoding='cp949')
```

```python
## str -> datetime으로 변환
v2 = file2['creationTime'].tolist()
date = []
date1 = []
for v in v2:
    dt = datetime.datetime.strptime(v, '%m/%d/%Y %H:%M')
    date.append(dt)
    
    dt2 = dt.strftime('%Y-%m-%d %H:%M')
    date1.append(dt2)
file2['creationTime'] = date
file2['creationDate'] = date1
```

### file3 load

```python
file3 = pd.read_csv('./data/isee_quality_data_20180901_to_201902.csv')
file3['qty_yn'] = [0 if x==1 else 1 for x in file3['Qty']]

### str -> Datetime 변환
v3 = file3['CreationTime'].tolist()
date = []
date1 = []
for v in v3:
    dt = datetime.datetime.strptime(v, '%m/%d/%Y %H:%M')
    date.append(dt)
    
    dt1 = dt.strftime('%Y-%m-%d')   # %Y-%m-%d %H:%M
    date1.append(dt1)
file3['creationTime'] = date
file3['creationDay'] = date1
```

### file 2, 3 merge

```python
file23 = pd.merge(file2, file3, how='left', on='creationDate')
```

```python
import seaborn as sns
```

```python
x = 10    # divide number
numOfgraphs = int(np.ceil(len(np.unique(isee_machine['CycleNum']))/x))
fig = plt.figure(figsize=(50,100), facecolor='white')
plt.subplots_adjust(hspace=1)

for i in range(numOfgraphs):
    df = isee_machine[(isee_machine['CycleNum']>=x*i) &(isee_machine['CycleNum']<=x*(i+1))]
    plt.subplot(numOfgraphs, 1, i+1)
    color_palette = sns.color_palette('hls', len(np.unique(df['CycleNum'])))
    
    sns.set(font_scale=2)
    g= sns.lineplot(data=df, x='creationDate', y=1, hue='CycleNum', legend='full', palette=color_palette)
    #g.set_xticklabels(g.get_xticklabels(), rotation=45)
    g.legend(loc='lower center', bbox_to_anchor=(0.5, -0.75), ncol=20)
```
