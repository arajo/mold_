```python
file1 = pd.read_excel('./data/C23_product_mapping_201809_201903.xlsx')
file1 = file1.dropna(axis=0)

file1['year'] = [str(x)[:4] for x in file1['生產日期']]
file1['month'] = [str(x)[5:7] for x in file1['生產日期']]
file1['day'] = [str(x)[8:10] for x in file1['生產日期']]
file1['count'] = 1 
file1
```

# C23_product
- day_list: unique한 일자가 담긴 list
- ind_list: unique한 일자의 인덱스 list
- new_file1: 하루에 한 개의 제품만 나오게!

```python
day_list = []    
ind_list = []    
for x in file1.index:
    if file1['生產日期'][x] not in day_list:
        day_list.append(file1['生產日期'][x])
        ind_list.append(x)

new_file1 = file1.loc[ind_list] 
```

```python
file3 = pd.read_csv('./data/isee_quality_data_20180901_to_201902.csv')
#file3['qty_yn'] = ['N' if x==1 else 'Y' for x in file3['Qty']]
file3['qty_yn'] = 'Y'

### str -> Datetime 변환
v3 = file3['CreationTime'].tolist()
date = []
date1 = []
for v in v3:
    dt = datetime.datetime.strptime(v, '%m/%d/%Y %H:%M')
    date.append(dt)
    
    dt1 = dt.strftime('%H')   # %Y-%m-%d %H:%M
    date1.append(dt1)
file3['creationTime'] = date
file3['Time'] = date1

# am, pm 판별 컬럼 추가
file3['ampm'] = [x.strftime('%p') for x in file3['creationTime']]
```

```python
file3['year'] = [str(x)[:4] for x in file3['creationTime']]
file3['month'] = [str(x)[5:7] for x in file3['creationTime']]
file3['day'] = [str(x)[8:10] for x in file3['creationTime']]
file3['CNT'] = 1
```
